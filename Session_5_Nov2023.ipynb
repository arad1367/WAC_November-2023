### Filter data

```python
print(f'Original length of spotify data is: {len(spotify_df)}')

# Filter_by_name
filter_by_name = spotify_df[spotify_df['artist(s)_name'] == 'Taylor Swift']
filter_by_name
print(f'filtered by name data length is: {len(filter_by_name)}')

# Filter by year
filter_by_year = spotify_df[spotify_df.released_year > 2022]
filter_by_year
print(f'filtered by year, data length is: {len(filter_by_year)}')
```

```python
spotify_df.tail()
```

```python
spotify_df.groupby('artist(s)_name').mean()
```

### Missing value problem

```python
spotify_df.isna().sum()
```

```python
spotify_df['key']
```

```python
# Solution 1  -- > dropna
spotify_dropna_dataset = spotify_df.dropna(inplace = False)
print(f"Length original data: {len(spotify_df)}")
print(f"Length with dropna: {len(spotify_dropna_dataset)}")

# Solution 2  -- > fillna
spotify_fillna_dataset = spotify_df['key'].fillna("missing")
print(f"Length with fillna: {len(spotify_fillna_dataset)}")
```

```python
# How to find most frequency
spotify_df['key'].value_counts()
```

### Numpy and different ways to create numpy array

```python
# simple 1D array
simple_array = np.array([1, 2, 3])
simple_array, simple_array.ndim

# 2D array
array_2D = np.array([
    [1, 2],
    [3,4]
])
array_2D, array_2D.ndim

# np.zeros
zeros_array = np.zeros(shape=(3, 3))
zeros_array

# np.ones
ones_array = np.ones(shape=(4, 4), dtype=int)
ones_array

# np.arange
arange_array = np.arange(1, 100, 10)
arange_array

# np.linspace
linspace_array = np.linspace(1, 2, 20)
linspace_array

# np.random
random_array = np.random.random(size=(10, 4))
random_array

# np.random.rand
rand_array = np.random.rand(10, 5)
rand_array
```

### Matplotlib

### Time series plot

```python
time_df = pd.DataFrame(data=np.random.normal(size=(1000, 4)),
                       index = pd.date_range('01/01/2023', periods=1000),
                       columns = ['Bitcoin', 'XRP', 'Solana', 'Sandbox'])
time_df = time_df.cumsum()
plt.figure();
time_df.plot();
```
